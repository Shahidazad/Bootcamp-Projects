```python
# Training data
#input (temp , rain ,humidity)
inputs=np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
```

```python
# Target data (apple,oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119]], dtype='float32')
```

```python
# convt to tensor
inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)
```

```python
# weight and bias
w=torch.randn(2,3,requires_grad=True) # 2,rows 3 columns # matrix
b=torch.randn(2,requires_grad=True) # vector
print(w)
b

```

```python
def model(x):
  return x @  w.t() +b  # x @(multiple) w.t(weight transpose matrix) b(bias)(vector) 
```

```python
# prediction
preds = model(inputs)
print(preds)
```

```python
targets
```

```python
diff =preds  - targets
diff_sqr=diff**2
diff_sqr
```

```python
# MSR loss
def mse(t1,t2):
  diff = t1-t2
  return torch.sum(diff**2) / diff.numel()
```

```python
# compute loss
loss= mse(preds,targets)
loss    # very bad
```

```python
# compute gradient 

loss.backward()
```
